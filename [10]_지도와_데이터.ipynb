## 4. DataFrameì„ ì´ìš©í•œ ì—¬ëŸ¬ ê´€ê´‘ì§€ í‘œì‹œ

ì´ë²ˆì—ëŠ” **ê´€ê´‘ì§€ ë¦¬ìŠ¤íŠ¸ë¥¼ DataFrame í˜•íƒœë¡œ ì €ì¥**í•˜ê³ , `for` ë°˜ë³µë¬¸ìœ¼ë¡œ ì—¬ëŸ¬ ì§€ì ì„ í•œ ë²ˆì— ì§€ë„ ìœ„ì— í‘œì‹œí•´ ë³´ê² ìŠµë‹ˆë‹¤.

```python
# ì œì£¼ ê´€ê´‘ì§€ ì˜ˆì‹œ ë°ì´í„°
names = ['ìš©ë‘ì•”', 'ì„±ì‚°ì¼ì¶œë´‰', 'ì •ë°©í­í¬', 'ì¤‘ë¬¸ê´€ê´‘ë‹¨ì§€', 'í•œë¼ì‚°1100ê³ ì§€', 'ì°¨ê·€ë„']
addr = [
    'ì œì£¼ì‹œ ìš©ë‘ì•”ê¸¸ 15',
    'ì„œê·€í¬ì‹œ ì„±ì‚°ì ì„±ì‚°ë¦¬',
    'ì„œê·€í¬ì‹œ ë™í™ë™ 299-3',
    'ì„œê·€í¬ì‹œ ì¤‘ë¬¸ë™ 2624-1',
    'ì„œê·€í¬ì‹œ ìƒ‰ë‹¬ë™ ì‚°1-2',
    'ì œì£¼ì‹œ í•œê²½ë©´ ê³ ì‚°ë¦¬ ì‚° 117'
]
```

```python
# DataFrame ìƒì„±
df = pd.DataFrame({'ê´€ê´‘ì§€ëª…': names, 'ì£¼ì†Œ': addr})
print("ì´ˆê¸° DataFrame ë¯¸ë¦¬ë³´ê¸°:")
print(df.head(), "\n")
```

```python
# ì£¼ì†Œë¥¼ ìœ„ë„/ê²½ë„ë¡œ ë³€í™˜í•˜ì—¬ ì—´ ì¶”ê°€
gk.add_coordinates_to_dataframe(df, 'ì£¼ì†Œ')
```

```python
# ì¢Œí‘œ ë°ì´í„° í™•ì¸
print("ì¢Œí‘œ ì¶”ê°€ í›„ DataFrame:")
print(df.head(), "\n")
```

```python
# ì¢Œí‘¯ê°’ì„ ìˆ«ìë¡œ ë³€í™˜
df['decimalLatitude'] = pd.to_numeric(df['decimalLatitude'])
df['decimalLongitude'] = pd.to_numeric(df['decimalLongitude'])
print("ì¢Œí‘œ ë°ì´í„° íƒ€ì… í™•ì¸:")
print(df.dtypes[['decimalLatitude', 'decimalLongitude']], "\n")
```

```python
# ì§€ë„ ì¤‘ì‹¬ ê³„ì‚°
center = [df['decimalLatitude'].mean(), df['decimalLongitude'].mean()]
print(f"ì§€ë„ ì¤‘ì‹¬ ì¢Œí‘œ (í‰ê· ê°’): {center}")
```

```python
# ì§€ë„ ìƒì„±
map = folium.Map(location=center, zoom_start=10)
```

```python
# ê° ê´€ê´‘ì§€ë§ˆë‹¤ ë§ˆì»¤ & í…ìŠ¤íŠ¸ í‘œì‹œ
print("\n ì§€ë„ì— ë§ˆì»¤ ì¶”ê°€ ì¤‘...")
for i in range(len(df)):
    loc = [df.loc[i, 'decimalLatitude'], df.loc[i, 'decimalLongitude']]
    folium.Marker(location=loc, icon=folium.Icon(color='red', icon='star')).add_to(map)
    html_label = f'<div style="font-size:12px; color:blue;"><b>{df.loc[i, "ê´€ê´‘ì§€ëª…"]}</b></div>'
    folium.Marker(location=loc, icon=folium.DivIcon(html=html_label)).add_to(map)
    print(f"  - {df.loc[i, 'ê´€ê´‘ì§€ëª…']} ë§ˆì»¤ ì¶”ê°€ ì™„ë£Œ!")

print("\n ëª¨ë“  ê´€ê´‘ì§€ ë§ˆì»¤ ì¶”ê°€ ì™„ë£Œ!")
map
```

## 1ï¸âƒ£ ì§€ë„(Map) ê°ì²´ í™•ì¥ ì˜µì…˜

Foliumì˜ `folium.Map()`ì€ ë‹¨ìˆœí•œ ì§€ë„ ê°ì²´ë¥¼ ìƒì„±í•˜ëŠ” ê²ƒ ì™¸ì—ë„

| `control_scale` | ì§€ë„ì˜ í•˜ë‹¨ì— **ê±°ë¦¬ ìŠ¤ì¼€ì¼ë°”**ë¥¼ í‘œì‹œí• ì§€ ì—¬ë¶€ë¥¼ ê²°ì •í•©ë‹ˆë‹¤. |

```python
# ğŸŒ Folium ê¸°ë³¸ ì§€ë„ ì„¤ì • ì˜ˆì œ
import folium

# 1ï¸âƒ£ ì§€ë„ ì¤‘ì‹¬ ì¢Œí‘œ ì§€ì • â€” ì„œìš¸ ì¢…ë¡œêµ¬ ê·¼ì²˜
center = [37.57, 126.98]
print("ì§€ë„ ì¤‘ì‹¬ ì¢Œí‘œ:", center)

# 2ï¸âƒ£ ë‹¤ì–‘í•œ ì˜µì…˜ê³¼ ìŠ¤íƒ€ì¼ ì ìš©
map1 = folium.Map(
    location=center,           # ì¤‘ì‹¬ ì¢Œí‘œ
    zoom_start=13,             # í™•ëŒ€ ìˆ˜ì¤€ (1~18)
    tiles="Stamen Terrain",    # ì§€ë„ ìŠ¤íƒ€ì¼
    control_scale=True,        # í•˜ë‹¨ ìŠ¤ì¼€ì¼ë°” í‘œì‹œ
    min_zoom=8,                # ìµœì†Œ í™•ëŒ€ ìˆ˜ì¤€ ì œí•œ
    max_zoom=17                # ìµœëŒ€ í™•ëŒ€ ìˆ˜ì¤€ ì œí•œ
)
print("Folium ì§€ë„ ê°ì²´ ìƒì„± ì™„ë£Œ!")

# 3ï¸âƒ£ ì§€ë„ ì¶œë ¥
map1
```

| `icon` | `folium.Icon()` ê°ì²´ë¡œ ë§ˆì»¤ì˜ ìƒ‰ìƒÂ·ëª¨ì–‘ì„ ì§€ì • |
| `draggable` | `True`ë¡œ ì„¤ì • ì‹œ ì‚¬ìš©ìê°€ ë§ˆì»¤ë¥¼ ë“œë˜ê·¸í•  ìˆ˜ ìˆìŒ |

## 4ï¸âƒ£ ë°˜ë³µë¬¸(for)ì„ ì´ìš©í•œ ë‹¤ì¤‘ ë§ˆì»¤ í‘œì‹œ

`for` ë°˜ë³µë¬¸ê³¼ DataFrameì„ ì¡°í•©í•˜ë©´  
ìˆ˜ì‹­~ìˆ˜ë°± ê°œì˜ ì§€ì ì„ **ìë™ìœ¼ë¡œ ì§€ë„ì— í‘œì‹œ**í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

| ê°œë… | ì„¤ëª… |
|------|------|
| `df.iterrows()` | DataFrameì˜ ê° í–‰(row)ì„ ë°˜ë³µ ì²˜ë¦¬í•  ë•Œ ì‚¬ìš© |
| `iloc` / `loc` | ì¸ë±ìŠ¤ ë˜ëŠ” ë¼ë²¨ ê¸°ë°˜ìœ¼ë¡œ ê°œë³„ ì¢Œí‘œë¥¼ ì°¸ì¡° |
| `add_to(map)` | ë§Œë“¤ì–´ì§„ ë§ˆì»¤ë¥¼ í˜„ì¬ ì§€ë„ ê°ì²´ì— ì¶”ê°€ |

> ğŸ’¡ `for`ë¬¸ ë‚´ì—ì„œ ì¡°ê±´ë¬¸(`if`)ì„ í•¨ê»˜ ì‚¬ìš©í•˜ë©´,  
> ì˜ˆë¥¼ ë“¤ì–´ íŠ¹ì • ì¡°ê±´(`ê´€ê° ìˆ˜ > 100ë§Œ`)ì—ë§Œ ë‹¤ë¥¸ ìƒ‰ìƒì˜ ë§ˆì»¤ë¥¼ í‘œì‹œí•˜ëŠ” ê²ƒë„ ê°€ëŠ¥í•©ë‹ˆë‹¤.

---

```python
# ğŸ” ë°˜ë³µë¬¸ì„ ì´ìš©í•œ ì—¬ëŸ¬ ì§€ì  ë§ˆì»¤ í‘œì‹œ
import folium
import pandas as pd

# ì œì£¼ ê´€ê´‘ì§€ ì˜ˆì‹œ ë°ì´í„°
data = {
    "ì´ë¦„": ["ìš©ë‘ì•”", "ì„±ì‚°ì¼ì¶œë´‰", "ì •ë°©í­í¬", "ì¤‘ë¬¸ê´€ê´‘ë‹¨ì§€", "í•œë¼ì‚°1100ê³ ì§€", "ì°¨ê·€ë„"],
    "ìœ„ë„": [33.5163, 33.4583, 33.2458, 33.2554, 33.3595, 33.3040],
    "ê²½ë„": [126.5110, 126.9424, 126.5623, 126.4122, 126.5129, 126.1621]
}
df = pd.DataFrame(data)
print("ê´€ê´‘ì§€ ë°ì´í„°í”„ë ˆì„ ë¯¸ë¦¬ë³´ê¸°:")
print(df.head(), "\n")

# ì§€ë„ ì¤‘ì‹¬ ê³„ì‚°
center = [df["ìœ„ë„"].mean(), df["ê²½ë„"].mean()]
print("ì§€ë„ ì¤‘ì‹¬:", center)

# ì§€ë„ ê°ì²´ ìƒì„±
map4 = folium.Map(location=center, zoom_start=10)

# ë°˜ë³µë¬¸ì„ ì´ìš©í•´ ë§ˆì»¤ ì¶”ê°€
for i, row in df.iterrows():
    loc = [row["ìœ„ë„"], row["ê²½ë„"]]
    name = row["ì´ë¦„"]

    # ë§ˆì»¤ ìƒ‰ìƒ ë³€ê²½ ê·œì¹™ (ì§ìˆ˜í–‰/í™€ìˆ˜í–‰)
    color = "red" if i % 2 == 0 else "blue"

    folium.Marker(
        location=loc,
        popup=f"{name} (ìƒ‰ìƒ: {color})",
        icon=folium.Icon(color=color, icon="star")
    ).add_to(map4)

print("ëª¨ë“  ê´€ê´‘ì§€ ë§ˆì»¤ ì¶”ê°€ ì™„ë£Œ!")
map4
```

## 5ï¸âƒ£ CircleMarkerë¡œ ë°ì´í„°ì˜ â€œí¬ê¸°â€ í‘œí˜„

`CircleMarker`ëŠ” ì›ì˜ ë°˜ì§€ë¦„(radius)ì„ ìˆ˜ì¹˜ ë°ì´í„°ì— ë¹„ë¡€í•˜ë„ë¡ ì„¤ì •í•  ìˆ˜ ìˆì–´,  
**ê°’ì˜ í¬ê¸° ì‹œê°í™”(ì˜ˆ: ì¸êµ¬ìˆ˜, ë§¤ì¶œì•¡, ë°©ë¬¸ì ìˆ˜)** ì— ë§¤ìš° ìœ ìš©í•©ë‹ˆë‹¤.

| ì†ì„±ëª… | ì„¤ëª… |
|--------|------|
| `radius` | ì›ì˜ ë°˜ì§€ë¦„ (ë°ì´í„° ê°’ì— ë¹„ë¡€ì‹œí‚¬ ìˆ˜ ìˆìŒ) |
| `color` | ì›ì˜ í…Œë‘ë¦¬ ìƒ‰ìƒ |
| `fill` | ë‚´ë¶€ ìƒ‰ìƒ ì±„ì›€ ì—¬ë¶€ (ê¸°ë³¸ê°’: `True`) |
| `fill_color` | ë‚´ë¶€ ìƒ‰ìƒ ì§€ì • |
| `fill_opacity` | ë‚´ë¶€ ìƒ‰ìƒì˜ íˆ¬ëª…ë„ (0~1 ì‚¬ì´ ê°’) |
| `stroke` | í…Œë‘ë¦¬ì„  í‘œì‹œ ì—¬ë¶€ |

> ğŸ’¡ **ë°ì´í„° í¬ê¸°ë¥¼ ì‹œê°í™”í•  ë•Œ**  
> ë‹¨ìˆœíˆ ìˆ˜ì¹˜ë¥¼ ì§ì ‘ ë°˜ì§€ë¦„ìœ¼ë¡œ ì“°ë©´ ì›ì´ ë„ˆë¬´ ì»¤ì§€ë¯€ë¡œ,  
> ë³´í†µ `radius = ê°’ ** 0.5` ë˜ëŠ” `ê°’ / ì¼ì • ìƒìˆ˜` í˜•íƒœë¡œ ìŠ¤ì¼€ì¼ë§í•©ë‹ˆë‹¤.

---

```python
# ğŸ§® CircleMarkerë¡œ ê°’ì˜ í¬ê¸°ë¥¼ ë°˜ì˜í•œ ì§€ë„ ë§Œë“¤ê¸°
import folium
import pandas as pd

# í’ì† ë°ì´í„° ì˜ˆì‹œ (ê°„ë‹¨í•œ 6ê°œ ì§€ì )
df = pd.DataFrame({
    "ë„ì‹œ": ["ì„œìš¸", "ë¶€ì‚°", "ëŒ€êµ¬", "ê´‘ì£¼", "ì¸ì²œ", "ì¶˜ì²œ"],
    "ìœ„ë„": [37.57, 35.17, 35.87, 35.16, 37.45, 37.87],
    "ê²½ë„": [126.98, 129.07, 128.60, 126.85, 126.70, 127.73],
    "í’ì†": [10.5, 8.2, 12.8, 9.0, 6.3, 15.0]
})
print("í’ì† ë°ì´í„°:")
print(df, "\n")

# ì§€ë„ ì¤‘ì‹¬ ê³„ì‚°
center = [df["ìœ„ë„"].mean(), df["ê²½ë„"].mean()]
print("ì§€ë„ ì¤‘ì‹¬:", center)

# ì§€ë„ ìƒì„±
map5 = folium.Map(location=center, zoom_start=6)

# ê° ë„ì‹œì˜ í’ì†ì— ë¹„ë¡€í•œ ì›í˜• í‘œì‹œ
for i, row in df.iterrows():
    loc = [row["ìœ„ë„"], row["ê²½ë„"]]
    radius = row["í’ì†"] ** 0.8  # ê°’ì˜ í¬ê¸°ë¥¼ ì¡°ì •í•˜ì—¬ ì› í¬ê¸° ì¡°ì ˆ
    folium.CircleMarker(
        location=loc,
        radius=radius,          # ì›ì˜ ë°˜ì§€ë¦„
        color="red",            # í…Œë‘ë¦¬ ìƒ‰
        fill=True,              # ë‚´ë¶€ ìƒ‰ ì±„ì›€
        fill_color="red",
        fill_opacity=0.5,       # íˆ¬ëª…ë„
        popup=f"{row['ë„ì‹œ']} í’ì†: {row['í’ì†']} m/s"
    ).add_to(map5)
    print(f"{row['ë„ì‹œ']} ë§ˆì»¤ ì¶”ê°€ (í’ì†={row['í’ì†']})")

print("ëª¨ë“  CircleMarker ì¶”ê°€ ì™„ë£Œ!")
map5
```
