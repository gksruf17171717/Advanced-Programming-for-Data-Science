# ğŸ§© Section 6. ë‹¨ì–´ í¼ì¹˜ê¸° (explode) ë° ì „ì²˜ë¦¬

```python
# ì˜ëª» ì¸ì‹ëœ ë‹¨ì–´ êµì •
nouns[nouns == "ëª¨ì‹œ"] = "í‹°ëª¨ì‹œ"
nouns[nouns == "IMAX"] = "ì•„ì´ë§¥ìŠ¤"
nouns[nouns == "íŒŒíŠ¸3"] = "3í¸"

# 2ê¸€ì ì´ìƒ ë‹¨ì–´ë§Œ ë‚¨ê¸°ê¸°

# ì¶”ì¶œëœ ëª…ì‚¬(nouns)ë¥¼ í•˜ë‚˜ì˜ ì»¬ëŸ¼ìœ¼ë¡œ ê°€ì§„ ìƒˆ ë°ì´í„°í”„ë ˆì„ ìƒì„±
# ì˜ˆ: word ì»¬ëŸ¼ì— ["ì‚¬ìš´ë“œ", "ì›…ì¥", "ì•„ì´ë§¥ìŠ¤", ...] í˜•íƒœë¡œ ì €ì¥ë¨
df_word = pd.DataFrame({"word": nouns})

# ê° ë‹¨ì–´ì˜ ê¸€ì ìˆ˜ë¥¼ ê³„ì‚°í•´ì„œ len ì»¬ëŸ¼ì— ì €ì¥
# ì˜ˆ: "ì‚¬ìš´ë“œ" â†’ 3, "ë„" â†’ 1
df_word["len"] = df_word["word"].str.len()

# ê¸€ì ìˆ˜ê°€ 2 ì´ìƒì¸ ë‹¨ì–´ë§Œ í•„í„°ë§
# ì˜ë¯¸ ì—†ëŠ” í•œ ê¸€ì ë‹¨ì–´(ì˜ë¯¸ ì—†ìŒ, ì¡°ì‚¬ ë“±)ë¥¼ ì œê±°í•˜ê¸° ìœ„í•´ ì‚¬ìš©
df_word = df_word.query("len >= 2")

print("ì „ì²˜ë¦¬ ì™„ë£Œ! ë‹¨ì–´ ê°œìˆ˜:", len(df_word))
df_word.head()
```

# ğŸ“Š Section 7. ë‹¨ì–´ ë¹ˆë„ ê³„ì‚° ë° ë¶ˆí•„ìš” ë‹¨ì–´ ì œê±°

```python
# ê°™ì€ ë‹¨ì–´ ë“±ì¥ íšŸìˆ˜ ì§‘ê³„

# wordë³„ë¡œ ê·¸ë£¹í•‘í•˜ì—¬ ê°™ì€ ë‹¨ì–´ê°€ ëª‡ ë²ˆ ë“±ì¥í–ˆëŠ”ì§€ ì„¸ì–´ì¤Œ
# count() â†’ í˜„ì¬ ë‚¨ì•„ ìˆëŠ” ì—´(len)ì„ ê¸°ì¤€ìœ¼ë¡œ ë“±ì¥ íšŸìˆ˜ë¥¼ ê³„ì‚°í•¨
df_word = df_word.groupby("word", as_index=False).count()

# len ì»¬ëŸ¼ì˜ ì´ë¦„ì„ countë¡œ ë³€ê²½ (í•´ë‹¹ ìˆ«ìê°€ ë‹¨ì–´ ë“±ì¥ íšŸìˆ˜ì´ë¯€ë¡œ)
# sort_values() â†’ ë“±ì¥ íšŸìˆ˜ê°€ ë§ì€ ë‹¨ì–´ ìˆœìœ¼ë¡œ ë‚´ë¦¼ì°¨ìˆœ ì •ë ¬
df_word = df_word.rename(columns={"len": "count"}).sort_values("count", ascending=False)

# ë¶ˆí•„ìš” ë‹¨ì–´ ì œê±°
del_idx = df_word.loc[df_word.word.isin(["ì˜í™”", "í¸ì´", "ì˜í™”ê´€", "íŒŒíŠ¸"])].index
df_word = df_word.drop(index=del_idx)

print("ìƒìœ„ ë‹¨ì–´ 10ê°œ:")
print(df_word.head(10))
```

# ğŸ“ˆ Section 8. ìƒìœ„ 10ê°œ ë‹¨ì–´ ì‹œê°í™”

```python
import matplotlib.pyplot as plt   # â† ë°˜ë“œì‹œ í•„ìš”!

df_top10 = df_word.head(10).sort_values("count", ascending=True)

plt.figure(figsize=(7,5))
plt.barh(df_top10["word"], df_top10["count"], color="skyblue")
plt.title("ì˜í™” ë¦¬ë·° ìƒìœ„ ë‹¨ì–´ TOP 10")
plt.xlabel("ë¹ˆë„ìˆ˜")
plt.ylabel("ë‹¨ì–´")
plt.show()
```

# â˜ï¸ Section 9. ì›Œë“œí´ë¼ìš°ë“œ ê¸°ë³¸ ì‹œê°í™”

---

## âš™ï¸ WordCloud ì£¼ìš” ì˜µì…˜ ìš”ì•½

| ì˜µì…˜ | ì„¤ëª… | ì˜ˆì‹œ |
|------|------|------|
| **font_path** | í•œê¸€ í°íŠ¸ ê²½ë¡œ ì§€ì • | `/usr/share/fonts/.../NanumGothic.ttf` |
| **width, height** | ì´ë¯¸ì§€ í¬ê¸° | `width=600, height=600` |
| **background_color** | ë°°ê²½ìƒ‰ ì§€ì • | `'white'`, `'black'`, `'ivory'` |
| **colormap** | ìƒ‰ìƒ í…Œë§ˆ | `'plasma'`, `'cool'`, `'rainbow'` |
| **mask** | ëª¨ì–‘ ì§€ì • (numpy ì´ë¯¸ì§€) | `mask=img` |
| **max_words** | í‘œì‹œí•  ë‹¨ì–´ ìˆ˜ ì œí•œ | `max_words=100` |
| **max_font_size** | ê¸€ì ìµœëŒ€ í¬ê¸° | `max_font_size=120` |
| **prefer_horizontal** | ê°€ë¡œ ê¸€ì ë¹„ìœ¨ | `prefer_horizontal=0.8` |
| **random_state** | ê²°ê³¼ ê³ ì • (ì¬í˜„ì„± í™•ë³´) | `random_state=123` |

---

```python
from wordcloud import WordCloud

# ë‹¨ì–´-ë¹ˆë„ ë”•ì…”ë„ˆë¦¬ë¡œ ë³€í™˜

# df_wordë¥¼ ë”•ì…”ë„ˆë¦¬ í˜•íƒœë¡œ ë³€í™˜
# {"ë‹¨ì–´": ë“±ì¥íšŸìˆ˜} í˜•íƒœê°€ â†’ WordCloud ìƒì„± ì‹œ í•„ìš”í•œ ë°ì´í„° êµ¬ì¡°
# ì˜ˆ: {"ì‚¬ìš´ë“œ": 12, "ì•„ì´ë§¥ìŠ¤": 9, "ì›…ì¥": 7, ...}
dic_word = df_word.set_index("word").to_dict()["count"]

# ì›Œë“œí´ë¼ìš°ë“œ ìƒì„±
wc = WordCloud(
    font_path=path,          # í•œê¸€ í°íŠ¸ ê²½ë¡œ (malgun.ttf ë“±) â†’ í•œê¸€ ê¹¨ì§ ë°©ì§€
    background_color="white",# ë°°ê²½ìƒ‰
    width=600,               # ì´ë¯¸ì§€ ê°€ë¡œ í¬ê¸°
    height=600,              # ì´ë¯¸ì§€ ì„¸ë¡œ í¬ê¸°
    colormap="cool",         # ìƒ‰ìƒ í…Œë§ˆ
    max_words=100,           # ìµœëŒ€ ë‹¨ì–´ ê°œìˆ˜
    random_state=123         # ëœë¤ ì‹œë“œ ê³ ì • (í•­ìƒ ê°™ì€ ê²°ê³¼)
)

# ë‹¨ì–´ ë¹ˆë„ ê¸°ë°˜ìœ¼ë¡œ ì›Œë“œí´ë¼ìš°ë“œ ì´ë¯¸ì§€ ìƒì„±
img_wordcloud = wc.generate_from_frequencies(dic_word)

# ì›Œë“œí´ë¼ìš°ë“œ ì¶œë ¥
plt.figure(figsize=(10,10))  # ì¶œë ¥ í¬ê¸° ì„¤ì •
plt.axis("off")              # ì¶• ì œê±° (ê¹”ë”í•˜ê²Œ ë³´ì´ë„ë¡)
plt.imshow(img_wordcloud)    # ì›Œë“œí´ë¼ìš°ë“œ ì´ë¯¸ì§€ í‘œì‹œ
plt.show()
```

# ğŸŒˆ Section 10. ëª¨ì–‘(mask)ê³¼ ìƒ‰ìƒ(colormap) ì‘ìš©

## ğŸ–¼ï¸ ëª¨ì–‘(mask)
- íŠ¹ì • ì´ë¯¸ì§€ í˜•íƒœ(ë³„, í•˜íŠ¸, ë¡œê³  ë“±) ì•ˆì— ë‹¨ì–´ë¥¼ ì±„ìš´ë‹¤.  
- ê²€ì€ìƒ‰ ë¶€ë¶„ì´ ë‹¨ì–´ê°€ ì±„ì›Œì§ˆ ì˜ì—­, í°ìƒ‰ì€ ë¹„ì–´ ìˆëŠ” ê³µê°„ì´ë‹¤.

## ğŸ¨ ìƒ‰ìƒ(colormap)
- `'plasma'`: ë”°ëœ»í•œ ì£¼í™©~ë³´ë¼  
- `'cool'`: ì‹œì›í•œ ì²­ë¡~ë³´ë¼  
- `'inferno'`: ê°•ë ¬í•œ ë…¸ë‘~ê²€ì •  
- `'rainbow'`: ë‹¤ì–‘í•œ ìƒ‰ìƒ  
- `'Greys'`: í‘ë°± ëŒ€ë¹„

```python
from google.colab import files
import PIL, numpy as np

# Colabì—ì„œ íŒŒì¼ ì—…ë¡œë“œ ê¸°ëŠ¥ ì‹¤í–‰
# ì‚¬ìš©ìê°€ cloud.png íŒŒì¼ì„ ì§ì ‘ ì„ íƒí•´ì•¼ í•¨
# ì—…ë¡œë“œí•œ íŒŒì¼ì€ í˜„ì¬ ì‘ì—… ë””ë ‰í† ë¦¬ì— ì €ì¥ë¨
print("cloud.png (ëª¨ì–‘ ì´ë¯¸ì§€)ë¥¼ ì—…ë¡œë“œí•˜ì„¸ìš”.")
uploaded = files.upload()
```

```python
# ì—…ë¡œë“œí•œ cloud.png ì´ë¯¸ì§€ë¥¼ ë¶ˆëŸ¬ì˜¤ê³  RGBA(íˆ¬ëª… í¬í•¨)ë¡œ ë³€í™˜
icon = PIL.Image.open("cloud.png").convert("RGBA")

# PIL ì´ë¯¸ì§€ ê°ì²´ â†’ numpy ë°°ì—´ë¡œ ë³€í™˜ (ì›Œë“œí´ë¼ìš°ë“œ maskë¡œ ì‚¬ìš©í•˜ê¸° ìœ„í•¨)
img = np.array(icon)

wc = WordCloud(
    font_path=path,           # í•œê¸€ í°íŠ¸ ê²½ë¡œ (í•œê¸€ ê¹¨ì§ ë°©ì§€)
    background_color="white", # ë°°ê²½ìƒ‰
    mask=img,                 # cloud.png ì´ë¯¸ì§€ë¥¼ ì›Œë“œí´ë¼ìš°ë“œ ëª¨ì–‘(mask)ìœ¼ë¡œ ì ìš©
    colormap="plasma",        # ìƒ‰ìƒ í…Œë§ˆ ì„¤ì •
    width=600,                # ìº”ë²„ìŠ¤ ê°€ë¡œ í¬ê¸°
    height=600,               # ìº”ë²„ìŠ¤ ì„¸ë¡œ í¬ê¸°
    max_words=80,             # ìµœëŒ€ ë‹¨ì–´ ê°œìˆ˜
    random_state=42           # ëœë¤ ì‹œë“œ ê³ ì • (ê²°ê³¼ ì¬í˜„ ê°€ëŠ¥)
)

# ë‹¨ì–´ ë¹ˆë„ ë”•ì…”ë„ˆë¦¬(dic_word)ë¥¼ ê¸°ë°˜ìœ¼ë¡œ ì›Œë“œí´ë¼ìš°ë“œ ìƒì„±
img_wordcloud = wc.generate_from_frequencies(dic_word)

plt.figure(figsize=(10,10))  # ì¶œë ¥ í¬ê¸° ì„¤ì •
plt.axis("off")              # ì¶• ì œê±° (ê¹”ë”í•˜ê²Œ ë³´ì´ë„ë¡)
plt.imshow(img_wordcloud)    # ìƒì„±ëœ ì›Œë“œí´ë¼ìš°ë“œ ì´ë¯¸ì§€ í‘œì‹œ
plt.show()
```

# ğŸ” Section 6. ì—°ê´€ë¶„ì„ (Association Rule Mining)

---

## ğŸ§  ê°œë… ìš”ì•½
- ê³ ê°ì´ **í•¨ê»˜ êµ¬ë§¤í•œ í•­ëª©(Item)** ë“¤ì„ **íŠ¸ëœì­ì…˜ ë‹¨ìœ„**ë¡œ ë¬¶ëŠ”ë‹¤.  
- â€œBurritoë¥¼ ì‚° ê³ ê°ì´ Chipsë„ ìƒ€ì„ í™•ë¥ â€ê³¼ ê°™ì€ ê´€ê³„ë¥¼ ì°¾ëŠ”ë‹¤.

---

## âš™ï¸ ì‹¤ìŠµ ì ˆì°¨
1ï¸âƒ£ `Transaction`ë³„ `Item`ì„ ë¦¬ìŠ¤íŠ¸ë¡œ ë¬¶ê¸°  
2ï¸âƒ£ ë¦¬ìŠ¤íŠ¸ë¥¼ **ì›-í•« ì¸ì½”ë”©(One-Hot Encoding)** í˜•íƒœë¡œ ë³€í™˜  
3ï¸âƒ£ `apriori()`ë¡œ ë¹ˆë°œ í•­ëª© ì§‘í•© ìƒì„±  
4ï¸âƒ£ `association_rules()`ë¡œ ì—°ê´€ ê·œì¹™ ë„ì¶œ

```python
# ì¤‘ë³µ ì œê±° í›„ íŠ¸ëœì­ì…˜ë³„ ì•„ì´í…œ ë¦¬ìŠ¤íŠ¸ ìƒì„±
# 1) í•„ìš”í•œ ë‘ ì»¬ëŸ¼ë§Œ ì„ íƒ: Transaction(ì£¼ë¬¸ë²ˆí˜¸), Item(ë©”ë‰´ëª…)
# 2) drop_duplicates() â†’ ë™ì¼í•œ ì£¼ë¬¸ì—ì„œ ê°™ì€ ë©”ë‰´ê°€ ì—¬ëŸ¬ ë²ˆ ì í˜€ ìˆì–´ë„ ì¤‘ë³µ ì œê±°
#    ì˜ˆ: Transaction 1001ì— 'Coke'ê°€ 2ë²ˆ ìˆìœ¼ë©´ 1ë²ˆë§Œ ë‚¨ê¹€
temp = df[['Transaction', 'Item']].drop_duplicates()

# 3) Transaction(ì£¼ë¬¸ë²ˆí˜¸) ê¸°ì¤€ìœ¼ë¡œ ê·¸ë£¹í™”
# 4) í•´ë‹¹ ì£¼ë¬¸ì—ì„œ ë“±ì¥í•œ Itemë“¤ì„ list í˜•íƒœë¡œ ë¬¶ê¸°
#    ì˜ˆ: 1001 â†’ ['Coke', 'Burger', 'Fries']
temp = temp.groupby('Transaction')['Item'].apply(list)

# 5) íŠ¸ëœì­ì…˜ë³„ ì•„ì´í…œ ë¦¬ìŠ¤íŠ¸ ì¼ë¶€ ì¶œë ¥
print("íŠ¸ëœì­ì…˜ ì˜ˆì‹œ:")
print(temp.head())
```

# íŠ¸ëœì­ì…˜ ì˜ˆì‹œ í•´ì„í•˜ê¸°

ì•„ë˜ ì¶œë ¥ì€ **ê° íŠ¸ëœì­ì…˜(= í•œ ë²ˆì˜ ì£¼ë¬¸)** ì•ˆì—ì„œ ê³ ê°ì´ ì–´ë–¤ ë©”ë‰´ë“¤ì„ í•¨ê»˜ ì£¼ë¬¸í–ˆëŠ”ì§€ë¥¼ ë³´ì—¬ì¤€ë‹¤.

---

## ì–´ë–»ê²Œ í•´ì„í•˜ë©´ ë ê¹Œ?

### ğŸ”¹ Transaction 1
- í•œ ê³ ê°ì´ í•œ ë²ˆì˜ ì£¼ë¬¸ì—ì„œ **Chips and Fresh Tomato Salsa**, **Izze**, **Nantucket** ë“±ì„ í•¨ê»˜ ì£¼ë¬¸í•¨  
- ì¦‰, ì—¬ëŸ¬ ì•„ì´í…œì´ ë™ì‹œì— ë“±ì¥í•œ ì£¼ë¬¸

### ğŸ”¹ Transaction 2
- **Chicken Bowl** í•˜ë‚˜ë§Œ ì£¼ë¬¸í•œ ê°„ë‹¨í•œ ì¼€ì´ìŠ¤

### ğŸ”¹ Transaction 3
- **Chicken Bowl + Side of Chips**  
- ë‘ ë©”ë‰´ê°€ í•¨ê»˜ êµ¬ë§¤ëœ ì£¼ë¬¸

### ğŸ”¹ Transaction 4

- ê°™ì€ ê³ ê°ì´ ë™ì‹œì— ë‘ ë©”ë‰´ë¥¼ ì£¼ë¬¸í•œ ì‚¬ë¡€

### ğŸ”¹ Transaction 5
- **Steak Burrito + Chips and Guacamole**  
- ë¶€ë¦¬ë˜ + ê³¼ì¹´ëª°ë¦¬ ì¹© ì¡°í•© ì£¼ë¬¸

---

## í•œ ì¤„ ìš”ì•½
**ì´ ë¦¬ìŠ¤íŠ¸ëŠ” â€œê° ì£¼ë¬¸(Transaction)ì—ì„œ ê³ ê°ì´ ì–´ë–¤ ë©”ë‰´ë“¤ì„ í•¨ê»˜ êµ¬ë§¤í–ˆëŠ”ì§€â€ë¥¼ ë³´ì—¬ì£¼ëŠ” ë°ì´í„°ì´ë‹¤.**

ì¥ë°”êµ¬ë‹ˆ ë¶„ì„(Association Rule / Apriori)ì„ ìˆ˜í–‰í•  ë•Œ ê°€ì¥ í•µì‹¬ì´ ë˜ëŠ” í˜•íƒœì˜ ë°ì´í„°ì´ë‹¤.


```python
# ì›-í•« ì¸ì½”ë”©ìœ¼ë¡œ ë³€í™˜ (TransactionEncoder)
# TransactionEncoder()ëŠ” ë¦¬ìŠ¤íŠ¸ í˜•íƒœì˜ ì¥ë°”êµ¬ë‹ˆ ë°ì´í„°ë¥¼
# â†’ ë¨¸ì‹ ëŸ¬ë‹/í†µê³„ ë¶„ì„ì— ì‚¬ìš©í•˜ê¸° ì¢‹ì€ 0/1 ë§¤íŠ¸ë¦­ìŠ¤ë¡œ ë³€í™˜í•˜ëŠ” ë„êµ¬
te = TransactionEncoder()

# te.fit(temp) : íŠ¸ëœì­ì…˜ì— ë“±ì¥í•˜ëŠ” ëª¨ë“  ì•„ì´í…œì„ í•™ìŠµ(ì•„ì´í…œ ëª©ë¡ ìƒì„±)
# te.transform(temp) : ê° íŠ¸ëœì­ì…˜ì„ 0/1 í˜•íƒœë¡œ ë³€í™˜
# ì˜ˆ: ['Steak Burrito', 'Chips'] â†’
#     Steak Burrito = 1, Chips = 1, ê·¸ ì™¸ ë©”ë‰´ = 0
trans_matrix = te.fit(temp).transform(temp)

# DataFrame ë³€í™˜
# numpy ë°°ì—´ í˜•íƒœë¡œ ë§Œë“¤ì–´ì§„ 0/1 ë§¤íŠ¸ë¦­ìŠ¤ë¥¼ DataFrameìœ¼ë¡œ ë³€í™˜
# columns=te.columns_ : ê° ì—´ ì´ë¦„ì„ 'ë©”ë‰´ ì´ë¦„'ìœ¼ë¡œ ì„¤ì •
basket = pd.DataFrame(trans_matrix, columns=te.columns_)

# ë³€í™˜ëœ ì¥ë°”êµ¬ë‹ˆ ë°ì´í„° ì•ë¶€ë¶„ í™•ì¸
# ê° í–‰ = í•˜ë‚˜ì˜ íŠ¸ëœì­ì…˜(ì£¼ë¬¸)
# ê° ì—´ = ë©”ë‰´ ì´ë¦„
# ê°’ 1 = í•´ë‹¹ ë©”ë‰´ê°€ ì£¼ë¬¸ë¨
# ê°’ 0 = ì£¼ë¬¸ë˜ì§€ ì•ŠìŒ
print("ë³€í™˜ ì™„ë£Œ! (0/1 ë§¤íŠ¸ë¦­ìŠ¤ êµ¬ì¡°)")
basket.head()
```

# âš¡ Section 7. Apriori ì•Œê³ ë¦¬ì¦˜ â€” ë¹ˆë°œ í•­ëª© ì§‘í•© ìƒì„±

---

## ğŸ’¬ Apriorië€?
- **Support(ì§€ì§€ë„)** ê°€ ì¼ì • ê¸°ì¤€ ì´ìƒì¸ í•­ëª© ì¡°í•©ë§Œ ë‚¨ê¸°ëŠ” ì•Œê³ ë¦¬ì¦˜  
- `min_support` ê°’ì„ ë‚®ì¶œìˆ˜ë¡ ë” ë§ì€ ì¡°í•©ì´ ë‚˜ì˜´ (ê³„ì‚°ëŸ‰ ì¦ê°€)

| íŒŒë¼ë¯¸í„° | ì„¤ëª… | ì˜ˆì‹œ |
|-----------|-------|------|
| **df** | ì›-í•« ì¸ì½”ë”©ëœ ë°ì´í„°í”„ë ˆì„ | basket |
| **min_support** | ìµœì†Œ ì§€ì§€ë„ | 0.01 â†’ 1% ì´ìƒ ê±°ë˜ì—ì„œ í•¨ê»˜ ë‚˜ì˜¨ í•­ëª© |
| **use_colnames** | ì‹¤ì œ ì•„ì´í…œ ì´ë¦„ í‘œì‹œ ì—¬ë¶€ | True |

---

```python
# Apriori ì‹¤í–‰
# apriori() : ì¥ë°”êµ¬ë‹ˆ(basket) ë°ì´í„°ì—ì„œ ìì£¼ ë“±ì¥í•˜ëŠ” ì•„ì´í…œ ì¡°í•©ì„ ì°¾ëŠ” í•¨ìˆ˜
# df=basket       â†’ ì›-í•« ì¸ì½”ë”©ëœ 0/1 ë§¤íŠ¸ë¦­ìŠ¤ë¥¼ ì…ë ¥
# min_support=0.01 â†’ ì „ì²´ íŠ¸ëœì­ì…˜ì˜ 1% ì´ìƒ ë“±ì¥í•˜ëŠ” ì¡°í•©ë§Œ ì°¾ìŒ
# use_colnames=True â†’ ê²°ê³¼ì— ì•„ì´í…œ ì´ë¦„(Item) ê·¸ëŒ€ë¡œ í‘œì‹œ (ìˆ«ì index ëŒ€ì‹ )
freq_item = apriori(df=basket, min_support=0.01, use_colnames=True)

print("[ë¹ˆë°œ í•­ëª© ì§‘í•©] ì˜ˆì‹œ:")
freq_item.head(10)
```

# ğŸ”— Section 8. ì—°ê´€ ê·œì¹™(Association Rules) ë„ì¶œ

---

## ğŸ“ˆ ì£¼ìš” ì§€í‘œ

| ì§€í‘œ | ì˜ë¯¸ | í•´ì„ ì˜ˆì‹œ |
|------|------|------------|
| **support** | ì „ì²´ ê±°ë˜ ì¤‘ í•´ë‹¹ ê·œì¹™ì´ í¬í•¨ëœ ë¹„ìœ¨ | 0.05 â†’ 5%ì˜ ê±°ë˜ì—ì„œ A, B ë™ì‹œ êµ¬ë§¤ |
| **confidence** | Aë¥¼ ìƒ€ì„ ë•Œ Bë„ í•¨ê»˜ ì‚° í™•ë¥  | 0.8 â†’ A êµ¬ë§¤ìì˜ 80%ê°€ Bë„ êµ¬ë§¤ |
| **lift** | A, Bê°€ ë…ë¦½ì¼ ë•Œ ëŒ€ë¹„ ëª‡ ë°° ë” ìì£¼ ë“±ì¥ | 1ë³´ë‹¤ í¬ë©´ ê¸ì •ì  ì—°ê´€ |

---

```python
# ì—°ê´€ ê·œì¹™ ë„ì¶œ
# association_rules() : Aprioriì˜ ë¹ˆë°œí•­ëª©ì§‘í•©(freq_item)ìœ¼ë¡œë¶€í„° ì—°ê´€ê·œì¹™ì„ ìƒì„±í•˜ëŠ” í•¨ìˆ˜
# df=freq_item  â†’ Apriorië¡œ ì°¾ì€ ì•„ì´í…œ ì¡°í•©ì„ ì…ë ¥
# metric='lift' â†’ ì—°ê´€ ê·œì¹™ì„ í‰ê°€í•  ê¸°ì¤€(metric)ìœ¼ë¡œ 'lift'ë¥¼ ì‚¬ìš©
# min_threshold=1 â†’ lift ê°’ì´ 1 ì´ìƒì¸ ê·œì¹™ë§Œ í•„í„°ë§ (1 ë¯¸ë§Œì€ ì˜ë¯¸ê°€ ì•½í•¨)
#   â€» lift > 1 : ë‘ ìƒí’ˆì´ í•¨ê»˜ ë“±ì¥í•  í™•ë¥ ì´ "ìš°ì—° ì´ìƒ"ìœ¼ë¡œ ë†’ë‹¤ (ê°•í•œ ê´€ë ¨ì„±)
rules = association_rules(df=freq_item, metric='lift', min_threshold=1)

# ì‹ ë¢°ë„ ë†’ì€ ìˆœìœ¼ë¡œ ì •ë ¬
# confidence(ì‹ ë¢°ë„) : Aë¥¼ ì‚¬ë©´ Bë„ ì‚´ í™•ë¥ 
# ì¦‰ â€œì–¼ë§ˆë‚˜ ì¼ê´€ì„± ìˆê²Œ í•¨ê»˜ íŒ”ë¦¬ëŠ”ê°€?â€ë¥¼ ë³´ì—¬ì£¼ëŠ” ì§€í‘œ
# ë†’ì€ confidenceì¼ìˆ˜ë¡ ë” ì‹ ë¢°í•  ìˆ˜ ìˆëŠ” ê·œì¹™
rules = rules.sort_values('confidence', ascending=False)

# ìƒìœ„ 10ê°œ ê·œì¹™ ì¶œë ¥
# â†’ ì–´ë–¤ ë©”ë‰´ ì¡°í•©ì´ í•¨ê»˜ ë§ì´ íŒ”ë¦¬ëŠ”ì§€ í™•ì¸ ê°€ëŠ¥
print("ìƒìœ„ 10ê°œ ì—°ê´€ ê·œì¹™:")
rules.head(10)
```

## ğŸ“˜ ì •ì œ ê·œì¹™
1ï¸âƒ£ ê¸€ì ìˆ˜ê°€ **2ê°œ ë¯¸ë§Œ**ì¸ ë‹¨ì–´ëŠ” ì œê±°  
2ï¸âƒ£ ì˜ë¯¸ ì—†ëŠ” ì¡°ì‚¬ë‚˜ ë‹¨ì–´ ì œê±° ê°€ëŠ¥ (`is`, `ê²ƒ`, `ìˆ˜` ë“±)

# ğŸ“Š Section 5. ë‹¨ì–´ ë¹ˆë„ ê³„ì‚° ë° ì •ë ¬

```python
# ë‹¨ì–´ë³„ ë“±ì¥ íšŸìˆ˜ ì§‘ê³„
# 1) 'word' ì»¬ëŸ¼ ê¸°ì¤€ìœ¼ë¡œ ë™ì¼í•œ ë‹¨ì–´ë“¤ì„ ë¬¶ìŒ
# 2) .count() â†’ ê° ë‹¨ì–´ê°€ ëª‡ ë²ˆ ë“±ì¥í–ˆëŠ”ì§€ ì„¸ì–´ 'count' ê°’ìœ¼ë¡œ ì €ì¥
#    ì˜ˆ: "ì‚¬ìš´ë“œ" 12ë²ˆ ë“±ì¥ â†’ count = 12
df_word = df_word.groupby('word', as_index=False).count() # groupby í›„ì—ë„ ê·¸ë£¹ ê¸°ì¤€ ì»¬ëŸ¼ì„ ì¸ë±ìŠ¤ë¡œ ì“°ì§€ ë§ê³ , ì¼ë°˜ ì»¬ëŸ¼ìœ¼ë¡œ ìœ ì§€í•´ë¼.

# ë“±ì¥ íšŸìˆ˜(count)ê°€ ë§ì€ ìˆœ(ë‚´ë¦¼ì°¨ìˆœ)ìœ¼ë¡œ ì •ë ¬
# ê°€ì¥ ë§ì´ ë“±ì¥í•œ ë‹¨ì–´ë¥¼ ìœ„ìª½ì— ë°°ì¹˜
df_word = df_word.sort_values('count', ascending=False)

print("ìƒìœ„ 10ê°œ ë‹¨ì–´:")
df_word.head(10)
```
